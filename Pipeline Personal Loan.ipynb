```python
#dropping the zip code 9307
loan_copy = loan_copy[loan_copy['ZIP Code'] != 9307]
```

```python
loan_copy.shape
```

```python
loan_copy.columns
```

```python
#ID can be dropped as it is a sequence number
#Experience can be removed because of the correlation with Age
X = loan_copy.drop(['ID', 'Experience', 'ZIP Code', 'Personal Loan'], axis = 1)
y = loan_copy['Personal Loan']
```

Split the data into train and test

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.25,random_state = 100)
```

APPLY LOGISTICREGRESSION ALGORITHM

```python
minmaxscaler = MinMaxScaler()
train_x = minmaxscaler.fit_transform(train_x)
test_x = minmaxscaler.fit_transform(test_x)

logisticregression = LogisticRegression(random_state = 1, solver = 'liblinear')
logisticregression.fit(train_x, train_y)
predict_y = logisticregression.predict(test_x)
print('The accuracy of the train:', logisticregression.score(train_x, train_y))
print('The accuracy of the test:', logisticregression.score(test_x, test_y))
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.25,random_state = 100)
```

```python
#The above steps are executed with the help of pipeline
pipe = Pipeline([("scaler",MinMaxScaler()), ("logisticregression", LogisticRegression(random_state=100, solver='liblinear'))])
pipe.fit(train_x, train_y)
predict_y = pipe.predict(test_x)
print("The accuracy score of the train:", pipe.score(train_x, train_y))
print("The accuracy score of the test:", pipe.score(test_x, test_y))

print("\nThe recall score of the model:", recall_score(test_y, predict_y))
print("\nThe confusion matrix of the model:", confusion_matrix(test_y, predict_y))
```

```python
pipe = make_pipeline(MinMaxScaler(), LogisticRegression(random_state=200, solver='liblinear'))
pipe.fit(train_x, train_y)
predict_y = pipe.predict(test_x)
print('The accuracy score of train:', pipe.score(train_x, train_y))
print('The accuracy score of test:', pipe.score(test_x, test_y))
```
